## Applying learned functions

```python
#import libraries
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({"Name": ['bread', 'beer', 'butter','cheese'],
                     "Type": ["bakery", "drinks", "deli", "deli"],
                     "Price": [2,5,3,6]})
```

```python
data
```

```python
#applying lambda function
data['inflated_price'] = data['Price'].apply(lambda x: x*1.4)

data
```

```python
#creating a flag column
data['price_label'] = data['inflated_price'].apply(
    lambda x: 'expensive' if x > 6 else ('acceptable' if x >=4 else 'cheap'))
data
```
